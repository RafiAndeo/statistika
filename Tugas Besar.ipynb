```
ggplot(data.baru, aes(x = Customer_Age, y = Attrition_Flag, color = Education_Level)) +
  geom_boxplot()
```

**Keterangan**

Pengguna yang memiliki umur kisaran 35 - 55 adalah nilai kuartil 1 sampai kuartil 3, dengan seluruh jenjang pendidikan dari mahasiswa sampai tidak diketahui. Hampir tidak ada pengguna yang berumur 70 tahun keatas namun berbeda dengan umur 30 kebawah. Di umur 30 tahun kebawah lebih banyak pengguna yang sudah tidak berlangganan lagi daripada pengguna yang masih berlangganan sama halnya dengan umur 70 tahun keatas.

```
data.baru %>%
  select(Credit_Limit, Total_Revolving_Bal) %>%
  ggpairs()
```

**Keterangan**

Nilai Korelasi dari limit kartu kredit dengan total saldo bergulir adalah 0.04249261. Nilai Korelasi yang tidak sama dengan 0 memiliki arti bahwa terdapat sebuah korelasi = H1. Tingkat kepercayaan yang hipotesis ini adalah 95% sedangkan tingkat kesalahan pada hipotesis ini adalah 5%.

```
cor.test(data.baru$Credit_Limit, data.baru$Total_Revolving_Bal)
```

*Kesimpulan*

Dari penyajian data di atas, dapat kita simpulkan bahwa pemegang kartu kredit yang pindah kebanyakan adalah yang berumur 40-50 tahun dan yang limit kartu kreditnya berada di kisaran 0-5.000 atau yang sekitar 35.000.

Dapat disimpulkan juga bahwa semakin banyak total pemegang kartu kredit di suatu kategori, semakin banyak juga yang pindah di kategori itu.
